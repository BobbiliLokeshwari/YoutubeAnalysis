```python
#import all necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('USvideos.csv')
```

```python
genre=pd.read_json('US_category_id (1).json')
```

```python
df.drop(['tags','thumbnail_link','description'],axis=1,inplace=True)
```

```python
df['trending_date']=pd.to_datetime(df['trending_date'],format='%y.%d.%m')
```

```python
df['pushlish_time']=pd.to_datetime(df['publish_time'])
```

```python
df.isnull().sum()
```

```python
genre['items'][0]
```

```python
id=[genre['items'][0]['id'] for i in range(len(genre))]
```

```python
gnr=[genre['items'][i]['snippet']['title'] for i in range(len(genre))]
```

```python
gr=pd.DataFrame({'category_id':id,'Genre':gnr})
```

```python
gr['category_id'] = gr['category_id'].astype(int)
```

```python
final_df=pd.merge(df,gr,on='category_id',how='inner')
```

```python
final_df
```

```python
final_df.info()
```

```python
final_df.drop(['video_id','trending_date','title','channel_title','publish_time','publish_time','pushlish_time'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
final_df['comments_disabled']=lb.fit_transform(final_df['comments_disabled'])
final_df['ratings_disabled']=lb.fit_transform(final_df['ratings_disabled'])
final_df['video_error_or_removed']=lb.fit_transform(final_df['video_error_or_removed'])
final_df['Genre_encode']=lb.fit_transform(final_df['Genre'])
```

```python
dict(zip(final_df['Genre_encode'], final_df['Genre']))

```

```python
final_df.drop(['category_id','Genre','ratings_disabled','video_error_or_removed','comments_disabled'],axis=1,inplace=True)
```

```python
final_df
```

MODEL BUILDING

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error as mse ,mean_absolute_error,r2_score
X=final_df.drop('likes',axis=1)
y=final_df['likes']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=30)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
err_rate=mse(y_test,y_pred)
score=r2_score(y_test,y_pred)
print('error:',err_rate)
print('rscore:',score)
```

```python
dr=DecisionTreeRegressor()
dr.fit(X_train,y_train)
y_pred=dr.predict(X_test)
err_rate=mse(y_test,y_pred)
score=r2_score(y_test,y_pred)
print('error:',err_rate)
print('rscore:',score)
```

```python
import pickle 
with open('model.pickle','wb') as model_file:
    pickle.dump(lr,model_file)
```
